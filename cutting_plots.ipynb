```python
def get_data(dat_label):
    retval = []
    
    graph_dirs = sorted(glob.glob('benchmark_results/dqva_{}/*'.format(dat_label)), key=lambda f: int(f.split('/')[-1].strip('G')))
    
    for graph_dir in graph_dirs:
        #print(graph_dir)
        all_reps = glob.glob(graph_dir + '/*rep*pickle')
        
        # Get the size of the optimal MIS
        gn = graph_dir.split('/')[-1]
        with open('benchmark_graphs/brute_force_outputs/N26_d3_graphs/{}_brute_force.out'.format(gn), 'r') as gf:
            for i, line in enumerate(gf):
                if i == 1:
                    opt = int(line.split()[-1])
                    break
        #print('\tOptimal MIS:', opt)
        
        for rep in all_reps:
            #print('\t'+rep)
    
            with open(rep, 'rb') as pf:
                out = pickle.load(pf)
                temp_mis = []
                for pround in out[1]:
                    if '0cuts' in dat_label:
                        best_indset, best_params, best_init_state, best_perm, partition, cut_nodes, history = pround
                    else:
                        best_indset, best_params, best_init_state, best_perm, partition, cut_nodes, hot_nodes, history = pround
                    
                    # Calculate and save the approximation ratio
                    temp_mis.append(hamming_weight(best_indset) / opt)
                #print('\t\tHamming weights:', temp_mis)
            retval.append(temp_mis)
    retval = np.array(retval)
    means = np.mean(retval, axis=0)
    stddevs = np.std(retval, axis=0)
    
    return (dat_label, (means, stddevs))
```

```python
all_data = []
for dat_label in ['0cuts', '1cuts', '2cuts', '3cuts']:
    print('Loading', dat_label)
    all_data.append(get_data(dat_label))
```

```python
fig, ax = plt.subplots(dpi=150, figsize=[6,3])

for label, dat in all_data:
    means, stddevs = dat
    x = np.arange(1, len(means)+1)
    real_label = 't = {}'.format(label.strip('cuts'))
    ax.errorbar(x, means, label=real_label, fmt='-o', ms=4)

    y1 = [min(m+s, 1) for m, s in zip(means, stddevs)]
    y2 = [m-s for m, s in zip(means, stddevs)]
    ax.fill_between(x, y1, y2, alpha=0.2)

plt.legend(loc='lower right', fontsize=12)
ax.set_xlabel('Partition Round', fontsize=14)
ax.set_ylabel('Approximation Ratio', fontsize=14)
ax.set_xticks([1,2,3,4])
plt.tight_layout()
#plt.savefig('figures/cutting_performance.pdf')
```

```python
bh_data = []

graph_dirs = sorted(glob.glob('benchmark_results/boppana-halldorsson/*'),
                              key=lambda f: int(f.split('/')[-1].strip('G')))
    
for graph_dir in graph_dirs:
    #print(graph_dir)
    all_reps = glob.glob(graph_dir + '/*rep*')

    # Get the size of the optimal MIS
    gn = graph_dir.split('/')[-1]
    with open('benchmark_graphs/brute_force_outputs/N26_d3_graphs/{}_brute_force.out'.format(gn), 'r') as gf:
        for i, line in enumerate(gf):
            if i == 1:
                opt = int(line.split()[-1])
                break
    #print('\tOptimal MIS:', opt)

    for rep in all_reps:
        #print('\t'+rep)
        with open(rep, 'r') as tf:
            for i, line in enumerate(tf):
                if i == 1:
                    mis = int(line.split()[-1])

        # Calculate and save the approximation ratio
        bh_data.append(mis / opt)

bh_data = np.array(bh_data)
print('Mean Boppana-Halldorsson approximation ratio:', np.mean(bh_data))
```
