```python
def load_optimal_mis(filename):
    with open(filename, 'r') as gf:
        for i, line in enumerate(gf):
            if i == 1:
                opt = int(line.split()[-1])
                return opt
```

```python
def get_data(graph_label):
    retval = []
    
    # Get cutting data
    all_labels = sorted(glob.glob(f'benchmark_results/{graph_label}/dqva_*cuts'),
                        key=lambda f: int(f.split('/')[-1].strip('dqva_cuts')))
    for dat_label in all_labels:
        #print('dat_label:', dat_label)
        num_cuts = int(dat_label.split('/')[-1].strip('dqva_cuts'))
        graph_dirs = sorted(glob.glob(f'{dat_label}/*'),
                            key=lambda f: int(f.split('/')[-1].strip('G')))

        temp_retval = []
        for graph_dir in graph_dirs:
            #print(graph_dir)
            all_reps = glob.glob(graph_dir + '/*rep*pickle')

            # Get the size of the optimal MIS
            gn = graph_dir.split('/')[-1]
            opt = load_optimal_mis('benchmark_graphs/brute_force_outputs/'
                                   f'{graph_label}_graphs/{gn}_brute_force.out')
            #print('\tOptimal MIS:', opt)
            all_rep_data = []
            for rep in all_reps:
                #print('\t'+rep)
                with open(rep, 'rb') as pf:
                    out = pickle.load(pf)
                temp_mis = []
                for pround in out[1]:
                    if num_cuts == 0:
                        best_indset, best_params, best_init_state, best_perm, partition, cut_nodes, history = pround
                    else:
                        best_indset, best_params, best_init_state, best_perm, partition, cut_nodes, hot_nodes, history = pround

                    # Calculate and save the approximation ratio
                    temp_mis.append(hamming_weight(best_indset) / opt)
                #print('\t\tHamming weights:', temp_mis)
                all_rep_data.append(temp_mis)

            temp_retval.append(sorted(all_rep_data, key=lambda l: l[-1])[-1])
        
        retval.append((num_cuts, temp_retval))

    retval = [(nc, np.array(val)) for nc, val in retval]
    final_cut_data = [(tup[0], np.mean(tup[1], axis=0), np.std(tup[1], axis=0)) for tup in retval]
    
    # Get Boppana-Halldorsson data
    bh_data = []
    graph_dirs = sorted(glob.glob(f'benchmark_results/boppana-halldorsson/{graph_label}_graphs/*'),
                        key=lambda f: int(f.split('/')[-1].strip('G')))
    
    for graph_dir in graph_dirs:
        #print(graph_dir)
        all_reps = glob.glob(graph_dir + '/*rep*')

        # Get the size of the optimal MIS
        gn = graph_dir.split('/')[-1]
        opt = load_optimal_mis('benchmark_graphs/brute_force_outputs/'
                               f'{graph_label}_graphs/{gn}_brute_force.out')
        #print('\tOptimal MIS:', opt)

        for rep in all_reps:
            #print('\t'+rep)
            with open(rep, 'r') as tf:
                for i, line in enumerate(tf):
                    if i == 1:
                        mis = int(line.split()[-1])

            # Calculate and save the approximation ratio
            bh_data.append(mis / opt)

    bh_data = np.array(bh_data)
    
    # Get Classical Divide and Conquer data
    dc_data = []
    graph_dirs = sorted(glob.glob('benchmark_results/classical_divide_and_conquer/'
                                  f'{graph_label}_graphs/*'),
                        key=lambda f: int(f.split('/')[-1].strip('G')))
    
    for graph_dir in graph_dirs:
        #print(graph_dir)
        all_reps = glob.glob(graph_dir + '/*rep*')

        # Get the size of the optimal MIS
        gn = graph_dir.split('/')[-1]
        opt = load_optimal_mis('benchmark_graphs/brute_force_outputs/'
                               f'{graph_label}_graphs/{gn}_brute_force.out')
        #print('\tOptimal MIS:', opt)
        all_rep_data = []
        for rep in all_reps:
            #print('\t'+rep)
            with open(rep, 'r') as tf:
                for i, line in enumerate(tf):
                    if i == 1:
                        mis = int(line.split()[-1].strip(']'))

            # Calculate and save the approximation ratio
            all_rep_data.append(mis / opt)
        dc_data.append(sorted(all_rep_data)[-1])

    dc_data = np.array(dc_data)

    return final_cut_data, np.mean(bh_data), np.mean(dc_data)
```

```python
dataset = 'N18_d3'
cut_data, avg_bh, avg_dc = get_data(dataset)
```

```python
fig, ax = plt.subplots(dpi=150, figsize=[6,3])

for label, means, stddevs in cut_data:
    x = np.arange(1, len(means)+1)
    real_label = f't = {label}'
    ax.errorbar(x, means, label=real_label, fmt='-o', ms=4)

    y1 = [min(m+s, 1) for m, s in zip(means, stddevs)]
    y2 = [m-s for m, s in zip(means, stddevs)]
    ax.fill_between(x, y1, y2, alpha=0.2)

ax.axhline(y=avg_bh, label='Boppana-Halldorsson', c='grey', alpha=0.6, ls='--')
ax.axhline(y=avg_dc, label='Classical D&C', c='purple', alpha=0.6, ls='--')

#plt.legend(loc='lower right', fontsize=12, ncol=3)
ax.set_xlabel('Partition Round', fontsize=15)
ax.set_ylabel('Approximation Ratio', fontsize=15)
ax.set_xticks(range(1,len(x)+1))
#ax.set_title(dataset)
#ax.set_xlim([0.8,8.2])
ax.set_ylim([0.5, 1.01])
t = ax.text(8, 0.524, f"3-regular, N = {dataset.strip('N_d3')}", ha="right", va="bottom", rotation=0, size=13,
    bbox=dict(boxstyle="round,pad=0.3", fc='w', ec='grey', alpha=0.4, lw=1))
plt.tight_layout()
#plt.savefig(f'figures/cutting-performance/{dataset}_full.pdf')
```
